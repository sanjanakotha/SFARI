```python
import pandas as pd
```

## TF AA coords

```python
SFARI_TFs = pd.read_csv("../data/SFARI_TFs_with_ENST.csv")
SFARI_TFs["End"] = SFARI_TFs["ProteinSeq"].str.len()
SFARI_TFs["Start"] = 1
SFARI_TFs
```

```python
# Adding ENSP codes
SFARI_TFs[["ENST"]].to_csv("../data/SFARI_tf_filtered_ENST_codes.txt", 
                                                 header=None, index=None, sep=' ', mode='a')
ENSP_codes = pd.read_csv("../data/mart_SFARI_TF_ENSP_output.txt", sep = "\t")
ENSP_codes
```

```python
SFARI_TFs_with_ENSP = pd.merge(SFARI_TFs, ENSP_codes, left_on = "ENST", 
         right_on = "Transcript stable ID version", how = "left")
SFARI_TFs_with_ENSP
```

```python
relevant_SFARI_TFs = SFARI_TFs_with_ENSP[["uniprotID", "Protein stable ID", "Start", "End"]]
relevant_SFARI_TFs
```

```python
relevant_SFARI_TFs = relevant_SFARI_TFs.rename(columns = {"Protein stable ID" : "ProteinID"})
relevant_SFARI_TFs
```

```python
relevant_SFARI_TFs.to_csv("../data/SFARI_TF_AA_coords.csv")
```
