#### Histogram of number of muts per TF

```python
plt.figure(figsize = (6,4), dpi = 300);
bins = np.arange(0, max(results["TF_missense"]) + 50, 50)
sns.histplot(results["TF_missense"] - results["AD_missense"], bins = bins, alpha = 0.5, label = "Rest of TF");
sns.histplot(results["AD_missense"], color = "orange", bins = bins, alpha = 0.5, label = "AD");
plt.legend()
plt.xlabel("Number of Missense SNPs")
plt.ylabel("TF Count");
```

```python
plt.figure(figsize = (6,4), dpi = 300);
min_val = min(min(results["TF_minus_AD_prop"]), min(results["AD_prop"]))
max_val = max(max(results["TF_minus_AD_prop"]), max(results["AD_prop"]))
bins = np.arange(min_val, max_val + 0.01, 0.01)
sns.histplot(results["TF_minus_AD_prop"], bins = bins, alpha = 0.5, label = "Rest of TF");
sns.histplot(results["AD_prop"], color = "orange", bins = bins, alpha = 0.5, label = "AD");
plt.legend()
plt.xlabel("Missense SNP Frequency")
plt.ylabel("Count");

```

```python
plt.figure(figsize = (6,4), dpi = 300);
min_val = min(min(results["TF_minus_AD_prop"]), min(results["AD_prop"]))
max_val = max(max(results["TF_minus_AD_prop"]), max(results["AD_prop"]))
bins = np.arange(min_val, max_val + 0.01, 0.01)
sns.kdeplot(results["TF_minus_AD_prop"], label = "Rest of TF");
sns.kdeplot(results["AD_prop"], color = "orange", label = "AD");
plt.legend()
plt.xlabel("Missense SNP Frequency")
plt.ylabel("Count");
```
