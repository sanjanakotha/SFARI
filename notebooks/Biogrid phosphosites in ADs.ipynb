## Looking for phosphorylation sites in ADs

Phosphosite sources
- Biogrid
- PhosphoSite Plus

1. Loading in all PTMs from Biogrid, keep phosphorylation sites
2. Match known ADs with this dataset using gene name and sequence

---
#### 1. Loading in all PTMs from Biogrid, keeping only phosphorylation sites

```python
biogrid_ptm_data = pd.read_csv("../data/BIOGRID-PTMS-4/BIOGRID-PTM-4.4.237.ptmtab.txt", sep = "\t", low_memory = False)
biogrid_ptm_data
```

```python
biogrid_ptm_data["Post Translational Modification"].value_counts()
```

```python
biogrid_phosphosites = biogrid_ptm_data[biogrid_ptm_data["Post Translational Modification"] == "Phosphorylation"]
biogrid_phosphosites
```

---
#### 2. Match known ADs with this dataset using gene name and sequence

```python
# loading in known ADs
known_ADs = pd.read_csv("../output/known_ADs_considering_isoforms_and_canonical.csv")
known_ADs
```

```python
# Merging phosphosites with ADs naively with gene name, then filtering to positions between AD start and end
merged_phosphosites_ADs = pd.merge(biogrid_phosphosites, known_ADs, left_on = "Official Symbol", right_on = "Gene", how = "inner")
merged_phosphosites_ADs = merged_phosphosites_ADs[merged_phosphosites_ADs["Position"].astype(int) < merged_phosphosites_ADs["End"]]
merged_phosphosites_ADs = merged_phosphosites_ADs[merged_phosphosites_ADs["Position"].astype(int) > merged_phosphosites_ADs["Start"]]
merged_phosphosites_ADs
```

```python
# Check that each AD starts and ends at the same place of full seq as expected 

isoform_match = []

for i in merged_phosphosites_ADs.index:
    
    full_seq = merged_phosphosites_ADs["Sequence"].loc[i]
    start = merged_phosphosites_ADs["Start"].loc[i] 
    end = merged_phosphosites_ADs["End"].loc[i] 
    ad_seq = merged_phosphosites_ADs["ProteinRegionSeq"].loc[i]


    exp_ad_seq = full_seq[start-1:end]
    isoform_match.append(ad_seq == exp_ad_seq)

merged_phosphosites_ADs["isoform_match"] = isoform_match
merged_phosphosites_ADs
```

```python
# Only keep rows where AD falls in start/end of full sequence as expected
isoform_matched_phosphosites = merged_phosphosites_ADs[merged_phosphosites_ADs["isoform_match"]]
isoform_matched_phosphosites
```

```python
# Only keep phosphosites on TFs
isoform_matched_phosphosites = isoform_matched_phosphosites[isoform_matched_phosphosites["TileType"] == "TF"]
isoform_matched_phosphosites
```

```python
isoform_matched_phosphosites.to_csv("../output/ad_biogrid_phosphorylation_sites.csv")
```
