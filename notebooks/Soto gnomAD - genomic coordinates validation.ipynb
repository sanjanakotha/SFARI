```python
SFARI_ADs = pd.read_csv("../output/SFARI_TF_known_ADs.csv", index_col = 0)
SFARI_ADs
```

```python
# for i in SFARI_ADs[SFARI_ADs["uniprotID"] == "P10275"].index:
for i in SFARI_ADs.index:
    uniprotID = SFARI_ADs["uniprotID"].loc[i]
    # uniprotID = "P10275"
    start = SFARI_ADs["Start"].loc[i]
    end = SFARI_ADs["End"].loc[i]
    
    print(uniprotID + "|" + str(start) + " - " + str(end) + "|" + uniprotBedTools.return_ENST(uniprotID))

    correct_seq = SFARI_ADs["ProteinSeq"].loc[i]
    found_seq = uniprotBedTools.AAseq_using_genomic_coords(uniprotID, start, end)
    
    if correct_seq != found_seq:
        print("MISMATCH!")
        # display(SFARI_ADs.loc[i, :])
        print("correct:")
        print(correct_seq)
        print("found:")
        print(found_seq)
    else:
        print("it's a match!")
    print("----")
```

```python
# for i in SFARI_ADs[SFARI_ADs["uniprotID"] == "P10275"].index:
for i in SFARI_TFs.index:
    uniprotID = SFARI_TFs["uniprotID"].loc[i]

    print(uniprotID + "|" + uniprotBedTools.return_ENST(uniprotID))

    correct_seq = SFARI_TFs["ProteinSeq"].loc[i]
    found_seq = uniprotBedTools.AAseq_using_genomic_coords(uniprotID, 1, len(correct_seq))
    
    if correct_seq != found_seq:
        print("MISMATCH!")
        # display(SFARI_ADs.loc[i, :])
        print("correct:")
        print(correct_seq)
        print("found:")
        print(found_seq)
    else:
        print("it's a match!")
    print("----")
```
