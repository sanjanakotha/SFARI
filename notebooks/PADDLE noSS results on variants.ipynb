```python
PADDLE = pd.read_csv('../output/PADDLE_noSS_SFARI_Variants_output.csv', index_col = 0)
PADDLE
```

PADDLE paper says "High-strength ADs were defined as the union of at least 5 consecutive 53-aa tiles with a predicted Z-score greater than 6. ADs defined in this way that overlapped by more than 26 aa were combined. Medium-strength ADs were defined similarly with a Z-score cutoff of 4 and were required to not overlap with high-strength ADs by more than 26 aa."

I will match up wild type and variant tiles. Will save variants that change prediction of tile.

```python
def get_start(AAseq,tiled_seq):
    return AAseq.index(tiled_seq)

PADDLE['tile_start'] = PADDLE.apply(lambda x: get_start(x.AAseq, x.tiled_sequences), axis=1)
```

```python
PADDLE["tile_len"] = PADDLE["tiled_sequences"].str.len()
PADDLE["tile_end"] = PADDLE['tile_start'] + PADDLE["tile_len"]
PADDLE
```

```python
PADDLE["AD"] = PADDLE["id"].str.split("|").str[0]
PADDLE["AD"] = PADDLE["AD"].replace({"NFE2L3 / NF2L3_Q9Y4A8_AD_252-451" : 'NFE2L3-NF2L3_Q9Y4A8_AD_252-451'})
PADDLE
```

```python
AD_set = set(PADDLE["AD"])
AD_set
```

```python
def return_longest_true_consec(lst):
    max_len = 0
    for k, g in groupby(lst):
        if k == True:
            max_len = max(max_len, len(list(g)))
    return max_len

def contains_AD(AD_rows):
    lst = AD_rows["active_tile"]
    if return_longest_true_consec(lst) >= 5:
        return True
    else:
        return False
```

```python
all_results = []

for AD in AD_set:
    AD_rows = PADDLE[PADDLE["AD"] == AD]
    AD_rows["active_tile"] = AD_rows["paddle_noSS_pred"] > 4
    
    # Separating PADDLE output rows into wt and variants of this AD
    
    wt_AD_rows = AD_rows[~AD_rows["id"].str.contains(">")]
    
    var_AD_rows = AD_rows[AD_rows["id"].str.contains(">")]
    var_AD_rows["var"] = var_AD_rows["id"].str.split("|").str[1]
    
    # For each variant, looking to see if it contains at least 1 predicted active region
    unique_vars = list(set(var_AD_rows["var"]))
    contains_AD_list = []

    for var_name in unique_vars:
        single_var_rows = var_AD_rows[var_AD_rows["var"] == var_name]
        contains_AD_list.append(contains_AD(single_var_rows))

    # Saving results 
    results = pd.DataFrame({"var" : unique_vars, "var_pred" : contains_AD_list})
    results["AD"] = AD
    results["orig_pred"] = contains_AD(wt_AD_rows)

    all_results.append(results)

all_results_df = pd.concat(all_results)
all_results_df
```
