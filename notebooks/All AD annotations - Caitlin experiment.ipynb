```python
clinvar_gene_filepaths= glob.glob("../data/clinvar_by_gene/*")
genes = pd.DataFrame({"filepath" : clinvar_gene_filepaths})
genes["gene"] = genes["filepath"].str.split("/").str[-1].str.split(".txt").str[0]
genes
```

```python
known_AD_15_entries = known_ADs[known_ADs["Gene"].isin(genes["gene"])]
known_AD_15_entries
```

```python
known_AD_15_entries["Reference"] = known_AD_15_entries["Reference"].str.split("//")
known_AD_15_entries
```

```python
known_AD_15_entries = known_AD_15_entries.explode("Reference")
known_AD_15_entries
```

```python
known_AD_15_entries["curr_list_len"] = known_AD_15_entries["End"] - known_AD_15_entries["Start"] + 1
known_AD_15_entries
```

```python
known_AD_15_entries["ref_start"] = known_AD_15_entries["Reference"].str.split("_").str[-2].str.split(" ").str[-1].astype(int)
known_AD_15_entries["ref_end"] = known_AD_15_entries["Reference"].str.split("_").str[-1].str.split(")").str[0].astype(int)
known_AD_15_entries["ref_list_len"] = known_AD_15_entries["ref_end"] - known_AD_15_entries["ref_start"] + 1
known_AD_15_entries
```

```python
known_AD_15_entries_summary = known_AD_15_entries[["Gene", "uniprotID", "Start", "End", "curr_list_len", "ref_list_len", "ref_start", "ref_end", "Reference"]]
known_AD_15_entries_summary
```

```python
known_AD_15_entries_summary.to_csv("../output/caitlin_experiment/all_annotations_in_15_ads.csv")
```
