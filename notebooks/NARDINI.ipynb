```python
import pandas as pd
from localcider.sequenceParameters import SequenceParameters
import time
```

```python
known_ADs = pd.read_csv("../output/known_ADs_considering_isoforms_and_canonical.csv")
known_ADs_on_TFs = known_ADs[known_ADs["TileType"] == "TF"]
known_ADs_on_TFs
```

```python
AD_score_matrices = []

for seq in known_ADs["ProteinRegionSeq"][:2]:
    print(seq)
    print(len(seq))
    t0 = time.time()
    SeqObj = SequenceParameters(seq)
    AD_score_matrices.append(SeqObj.calculate_zscore(num_scrambles=100000, random_seed=None))
    t1 = time.time()
    print(t1 - t0)
    print('--------------\n')
```
