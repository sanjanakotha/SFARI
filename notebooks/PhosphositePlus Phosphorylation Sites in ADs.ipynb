### Phosphosite Plus Phosphorylation sites in ADs
1. Downloaded dataset from Phosphosite website https://www.phosphosite.org/staticDownloads
2. Used uniprotID to match with known ADs
3. Verified that residue at position matches expectation

```python
phosphosite_plus = pd.read_csv("../data/Phosphorylation_site_dataset", header = 2, sep = "\t")
phosphosite_plus
```

```python
phosphosite_plus["resid"] = phosphosite_plus["MOD_RSD"].str[0]
phosphosite_plus["pos"] = phosphosite_plus["MOD_RSD"].str[1:].str.split("-").str[0].astype(int)
phosphosite_plus
```

```python
# Using uniprotID to match up with known ADs

known_ADs = pd.read_csv("../output/known_ADs_considering_isoforms_and_canonical.csv")
known_ADs
```

```python
uniprotID_merged = pd.merge(phosphosite_plus, known_ADs, left_on = "ACC_ID", right_on = "uniprotID")
uniprotID_merged
```

```python
# Check that start < position < end

pos_in_bounds = uniprotID_merged[(uniprotID_merged["Start"] < uniprotID_merged["pos"]) & (uniprotID_merged["pos"] < uniprotID_merged["End"])]
pos_in_bounds
```

```python
# check that right residue is at position as described by modification

for i in pos_in_bounds.index:

    # Check position
    row = pos_in_bounds.loc[i]
    AD_seq = row.ProteinRegionSeq
    pos = row.pos
    exp_resid = row.resid
    start = row.Start

    AD_pos = pos - start
    obs_resid = AD_seq[AD_pos]

    if (obs_resid != exp_resid):
        print("mismatch!")
        print(row.Gene)
```

```python
TF_pos_in_bounds = pos_in_bounds[pos_in_bounds["TileType"] == "TF"]
TF_pos_in_bounds
```
