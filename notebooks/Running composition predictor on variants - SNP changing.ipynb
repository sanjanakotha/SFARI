```python
import protfasta

# Number of wt ADs
wt_input = protfasta.read_fasta("../../SFARI/soto_analysis/data/wt_ADs.fasta")
wt_seqs = pd.DataFrame({"AD" : wt_input.keys(), "seq" : wt_input.values()})
wt_seqs
```

```python
# Number of wt ADs containing at least one predicted AD
wt_AD_results = pd.DataFrame({"AD" : wt_input.keys()})
wt_AD_results["contains_active_pred"] = wt_AD_results["AD"].isin(wt_preds["GeneName"])
wt_AD_results
```

```python
variant_preds = AD_predictor_tools.make_predictions(folder_name="predictions/",
                        inputfilename="../../SFARI/soto_analysis/data/long_variant_ADs.fasta", 
                        slope=1,
                        lower_corner_c="VP16",lower_corner_h=6,
                        upper_corner_c=-8,upper_corner_h="CITED2",
                        LowerCorner_slope1='inf', LowerCorner_slope2=0,
                        UpperCorner_slope1='inf', UpperCorner_slope2='inf',
                        composition=["W","F","L"],
                        window_size=39,
                        window_spacing=1,
                        propset=["Charge","AllHydros"])
```

```python
variant_preds
```

```python
# Number of wt AD variants
variant_input = protfasta.read_fasta("../../SFARI/soto_analysis/data/long_variant_ADs.fasta")
variant_input_df = pd.DataFrame({"AD_variant" : variant_input.keys()})
variant_input_df["variant_activity"] = variant_input_df["AD_variant"].isin(variant_preds["GeneName"])
variant_input_df
```

```python
variant_input_df["AD_variant_trunc"] = variant_input_df["AD_variant"].str.split("|").str[0]
variant_input_df
```

```python
merged = pd.merge(wt_AD_results, variant_input_df, left_on = "AD", right_on = "AD_variant_trunc")
merged
```

```python
mismatches = merged[(merged["contains_active_pred"]) != (merged["variant_activity"])]
mismatches
```

```python
changes = mismatches["AD_variant"].str.extract(r'p.([a-zA-Z]{3})\d+([a-zA-Z]{3})')
changes[0] = changes[0].str.upper()
changes[1] = changes[1].str.upper()
changes = changes.rename(columns = {0 : "orig", 1 : "var"})
changes = changes.reset_index(drop = True)
changes
```
