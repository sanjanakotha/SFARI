```python
lambert_TF_ENSTs = pd.read_csv("../output/TFs_with_ENST.csv", index_col = 0)

uniprotID_ENST_mapping_dict = dict(zip(lambert_TF_ENSTs["canonical_uniprot_id_isoform"],
                                       lambert_TF_ENSTs["ENST"]))

uniprotID_isoform_mapping_dict = dict(zip(lambert_TF_ENSTs["uniprotID"],
                                       lambert_TF_ENSTs["canonical_uniprot_id_isoform"]))
#uniprotID_ENST_mapping_dict
lambert_TF_ENSTs
```

```python
# From: http://localhost:8888/notebooks/Desktop/Staller_Lab/SFARI/notebooks/Looking%20for%20non-canonical%20ADs.ipynb
AD_coords = pd.read_csv("../output/known_ADs_considering_isoforms_and_canonical.csv", index_col = None)
AD_coords = AD_coords[AD_coords["TileType"] == "TF"]
AD_coords["ENSG"] = np.nan
AD_coords = AD_coords[["uniprotID", "Start", "End", "ENSG"]]
AD_coords["uniprotID"] = AD_coords["uniprotID"].map(uniprotID_isoform_mapping_dict)
AD_coords
```

```python
AD_coords["coord"] = AD_coords["Start"].astype(str)+"-"+AD_coords["End"].astype(str)
AD_coords
```

```python
merged_AD_coords = AD_coords[["uniprotID", "coord"]].groupby("uniprotID").agg(lambda c: ",".join(c))
#merged_AD_coords = merged_AD_coords.drop(["O95718"])
merged_AD_coords = merged_AD_coords.reset_index()
merged_AD_coords
```

```python
TF_table = merged_AD_coords.rename(columns = {"coord" : "AD_coords"})
TF_table = pd.merge(TF_table, AD_coords[["uniprotID", "ENSG"]], on = "uniprotID", how = "left").drop_duplicates()
TF_table = TF_table.reset_index(drop = True)
TF_table
```

```python
TF_table["ENST"] = TF_table["uniprotID"].map(uniprotID_ENST_mapping_dict)
TF_table["DBD_coords"] = "NA-NA"
TF_table["RD_coords"] = ""
TF_table["Bif_coords"] = ""
TF_table["1"] = ""
TF_table["2"] = ""
TF_table["length"] = "1"
TF_table
```

```python
TF_table_formatted = TF_table[["1","2", "uniprotID", "ENSG", "ENST", "DBD_coords", "AD_coords", "RD_coords", "Bif_coords", "length"]]
TF_table_formatted
```

```python
len(TF_table_formatted[TF_table_formatted["ENST"].isna()])
```

```python
lambert_TF_DBDs = pd.read_csv("../output/lambert_TFs_10-21-24_with_DBD_coords.csv")
lambert_TF_DBDs["uniprotID"] = lambert_TF_DBDs["id"].str.split("|").str[1]
lambert_TF_DBDs["uniprotID"] = lambert_TF_DBDs["uniprotID"].map(uniprotID_isoform_mapping_dict)
lambert_TF_DBDs = lambert_TF_DBDs[["uniprotID", "DBD_coords_merged"]]
lambert_TF_DBDs
```

```python
TF_table_formatted = pd.merge(TF_table_formatted, lambert_TF_DBDs, how = "left")
TF_table_formatted
```

```python
import ast

def convert_ranges_from_string(input_string):
    # Convert the string to a list of lists
    input_list = ast.literal_eval(input_string)
    
    # Convert the list of lists into the desired string format
    return ','.join([f"{start}-{end}" for start, end in input_list])
```

```python
DBD_coords = []
for i in TF_table_formatted["DBD_coords_merged"]:
    # no DBD
    if type(i) != str:
        DBD_coords.append("NA-NA")
    else:
        DBD_coords.append(convert_ranges_from_string(i))

TF_table_formatted["DBD_coords"] = DBD_coords
TF_table_formatted
```

```python
TF_table_formatted = TF_table_formatted.drop(columns = "DBD_coords_merged")
TF_table_formatted
```

```python
TF_table_formatted["uniprotID"] = TF_table_formatted["uniprotID"].str.split("-").str[0]
TF_table_formatted
```

```python
TF_table_formatted["ENST"] = TF_table_formatted["ENST"].str.split(".").str[0]
TF_table_formatted
```
