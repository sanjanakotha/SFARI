# Percent identities to compare conservation of TF groups

## 1. Calculating percent identity for each position of full protein alignment

```python
alignment_paths = set(glob.glob("../data/zoonomia_toga_mca/prot_alignments/*")) | set(glob.glob("../data/zoonomia_toga_mca/prot_alignments_wrong_isoforms/*"))
len(alignment_paths)
```

```python
for path in alignment_paths:
    gene = path.split("/")[-1].split(".")[0]
    alignment = AlignIO.read(path, "fasta")
    percent_identities = []
    summary_align = AlignInfo.SummaryInfo(alignment)
    
    for record in alignment:
        if "REFERENCE" in record.id:
            reference_seq = record.seq
    
    percent_identities = []
    positions = []
    
    # Iterate over each position in the alignment where ref_AA is not a gap
    for i in range(alignment.get_alignment_length()):
        ref_AA = reference_seq[i]

        # if ref_AA == "X" or ref_AA == "-":
        #     print(gene)

        if ref_AA != "-":
            # Get the column (position) for all sequences
            column = alignment[:, i]
        
            # Calculating percent identity
            count = column.count(ref_AA)
            percent_identity = 100 * count / len(column)
        
            # AAs.append(ref_AA)
            percent_identities.append(percent_identity)
            #positions.append(lam_AA)
        
    percent_identities_df = pd.DataFrame({"percent_identity" : percent_identities})
    percent_identities_df.to_csv("../data/zoonomia_toga_mca/prot_alignment_percent_identities/" + gene)
    #display(percent_identities_df)
```

## 2. Calculate and save average percent identity per TF

```python
genes = []
avg_perc_idents = []

for path in glob.glob("../data/zoonomia_toga_mca/prot_alignment_percent_identities/*"):
    gene = path.split("/")[-1].split(".")[0]
    df = pd.read_csv(path)
    
    avg_perc_idents.append(np.mean(df["percent_identity"]))
    genes.append(gene)
```

```python
avg_perc_ident_df = pd.DataFrame({"gene" : genes, "avg_percent_identity" : avg_perc_idents})
avg_perc_ident_df
```

```python
sns.set_context('talk')
sns.set_style('ticks')
sns.histplot(avg_perc_ident_df["avg_percent_identity"])
sns.despine()
plt.title("Activator TFs")
```

## 3. Finding overlap with SFARI TFs

```python
SFARI_genes = pd.read_csv("../data/SFARI-Gene_genes_01-23-2023release_03-21-2023export.csv")
SFARI_genes
```

```python
SFARI_activator_genes = avg_perc_ident_df[avg_perc_ident_df["gene"].isin(SFARI_genes["gene-symbol"])]
SFARI_activator_genes
```

```python
non_SFARI_activator_genes = avg_perc_ident_df[~avg_perc_ident_df["gene"].isin(SFARI_genes["gene-symbol"])]
non_SFARI_activator_genes
```

```python
sns.histplot(non_SFARI_activator_genes["avg_percent_identity"],stat = 'density', bins = np.arange(44, 101), edgecolor = 'none', label = "Other TFs")
plt.axvline(np.mean(non_SFARI_activator_genes["avg_percent_identity"]), color = sns.color_palette('tab10')[0], linestyle = "--")
           
sns.histplot(SFARI_activator_genes["avg_percent_identity"],stat = 'density', bins = np.arange(44, 101), edgecolor = 'none', label = "SFARI TFs")
plt.axvline(np.mean(SFARI_activator_genes["avg_percent_identity"]), color = sns.color_palette('tab10')[1], linestyle = "--")

plt.legend()
plt.xlabel("Average Percent Identity")
plt.title("Activator TFs")
sns.despine()
```
