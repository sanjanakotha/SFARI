```python
import pandas as pd
import glob
import numpy as np
import protfasta
```

- Patient variants
- AlphaMissense 10 most + least pathogenic variants per AD
- All gnomAD variants in AD
- All Clinvar variants in AD

```python
cc_names = pd.read_excel("../data/gene_names_with_location_and_pmvs.xlsx")
cc_names
```

```python
# Loading in the AD names
variant_fasta_filepaths = glob.glob('../output/caitlin_experiment/variant fastas/*')
ADs = pd.DataFrame({"filepath" : variant_fasta_filepaths})
ADs["gene"] = ADs["filepath"].str.split("fastas/").str[1].str.split("_").str[0]
ADs["uniprotID"] = ADs["filepath"].str.split("_").str[2]
ADs["start"] = ADs["filepath"].str.split("_AD_").str[1].str.split("-").str[0].astype(int)
ADs["end"] = ADs["filepath"].str.split("_AD_").str[1].str.split("-").str[1].str.split("_").str[0].astype(int)
ADs["name"] = ADs["filepath"].str.split("/").str[-1].str.split("_seq_adj").str[0]
ADs
```

```python
ADs = pd.merge(ADs, cc_names)
ADs
```

```python
sk_cc_name_dict = dict(zip(ADs["name"], ADs["Gene Name"]))
sk_cc_name_dict
```

```python
def return_var_df(filepath, colname, clinvar = False, alphamissense = False, south_asian = False):
    ADs_with_clinvar_vars = [_.split("/")[-1] for _ in glob.glob("../soto_analysis/outputs/AD_clinvar_for_15_variant_fasta/*")]
    ADs_with_am_vars = [_.split("/")[-1] for _ in glob.glob("../soto_analysis/outputs/AD_alphamissense_top10_variant_fasta/*")]
    ADs_with_south_asian_vars = [_.split("/")[-1] for _ in glob.glob("../soto_analysis/outputs/AD_south_asian_variant_fasta/*")]

    return_df = []
    
    for AD in set(ADs["name"]):
        if clinvar and AD not in ADs_with_clinvar_vars:
            return_df.append(pd.DataFrame())
        elif alphamissense and AD not in ADs_with_am_vars:
            return_df.append(pd.DataFrame())
        elif ADs_with_south_asian_vars and AD not in ADs_with_south_asian_vars:
            return_df.append(pd.DataFrame())
        else:
            fasta_dict = protfasta.read_fasta(filepath + AD)
            fasta_df = pd.DataFrame({"Gene Name": sk_cc_name_dict[AD], 
                                     "Mutant Sequence":fasta_dict.values(),
                                     "Protein Change": fasta_dict.keys()})
            if len(fasta_df) > 0:
                #display(fasta_df)
                fasta_df["Protein Change"] = fasta_df["Protein Change"].str.split("(").str[1].str.split(")").str[0]
                fasta_df = fasta_df.drop_duplicates(subset = ["Mutant Sequence"])
                return_df.append(fasta_df)
            else:
                return_df.append(pd.DataFrame())
        
    return_df = pd.concat(return_df)
    return_df["Mutation"] = colname
    return return_df.reset_index(drop = True)
```

```python
patient_vars = return_var_df("../soto_analysis/outputs/AD_variant_fasta/", "patient")
patient_vars.to_csv("../output/caitlin_experiment/patient_vars.csv")
patient_vars
```

```python
gnomad_vars = return_var_df("../soto_analysis/outputs/AD_gnomad_variant_fasta/", "gnomad")
gnomad_vars.to_csv("../output/caitlin_experiment/gnomad_vars.csv")
gnomad_vars
```

```python
clinvar_vars = return_var_df("../soto_analysis/outputs/AD_clinvar_for_15_variant_fasta/", "clinvar", clinvar = True)
clinvar_vars.to_csv("../output/caitlin_experiment/clinvar_vars.csv")
clinvar_vars
```

```python
am_top10_vars = return_var_df("../soto_analysis/outputs/AD_alphamissense_top10_variant_fasta/", "am_top10", alphamissense=True)
am_top10_vars.to_csv("../output/caitlin_experiment/am_top10_vars.csv")
am_top10_vars
```

```python
am_bottom10_vars = return_var_df("../soto_analysis/outputs/AD_alphamissense_bottom10_variant_fasta/", "am_bottom10", alphamissense=True)
am_bottom10_vars.to_csv("../output/caitlin_experiment/am_bottom10_vars.csv")
am_bottom10_vars
```

```python
south_asian_vars = return_var_df("../soto_analysis/outputs/AD_south_asian_variant_fasta/", "south_asian", south_asian=True)
south_asian_vars.to_csv("../output/caitlin_experiment/south_asian_vars.csv")
south_asian_vars
```
