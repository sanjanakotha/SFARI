```python
def split_coords(data, strand):
	# https://stackoverflow.com/questions/3149440/splitting-list-based-on-missing-numbers-in-a-sequence
	output = []
	if strand == "-" or strand == -1:
		data.reverse()
	for k, g in groupby(enumerate(data), lambda i_x: i_x[0] - i_x[1]):
		output.append(list(map(itemgetter(1), g)))
	return output
```

```python
#https://www.geeksforgeeks.org/python-make-a-list-of-intervals-with-sequential-numbers/
def interval_extract(list):
    list = sorted(set(list))
    range_start = previous_number = list[0]
 
    for number in list[1:]:
        if number == previous_number + 1:
            previous_number = number
        else:
            yield [range_start, previous_number]
            range_start = previous_number = number
    yield [range_start, previous_number]
 
```
