```python
### Check, does site +/- 7 AA exist in each ProteinRegionSeq

for eight_AAs, full in zip(cc_phosphosites["SITE_+/-7_AA"], cc_phosphosites["ProteinRegionSeq"]):
    if eight_AAs.upper() in full:
        print("good")
    else:

        print("MISMATCH")
        print(eight_AAs)
        print(full)

    print()
```

```python
cc_phosphosites
```

```python
# Switching all phosphosites to As
cc_phosphosites_to_a = cc_phosphosites.copy(deep = True)
cc_phosphosites_to_a
```

```python
cc_phosphosites_to_a.loc[0]
```

```python
new_seqs = []
prot_changes = []

for i in cc_phosphosites_to_a.index:
    row = cc_phosphosites_to_a.loc[i]
    site_AD_pos = row.pos - row.Start
    orig_seq = row.ProteinRegionSeq
    aa_dict = {"S" : "Ser",
               "T" : "Thr",
               "Y": "Tyr"}
    prot_changes.append("p." + aa_dict[orig_seq[site_AD_pos]] + str(row.pos) + "Ala")

    new_seq = orig_seq[:site_AD_pos] + "a" + orig_seq[site_AD_pos + 1:]
    # print(row.MOD_RSD)
    # print(orig_seq)
    # print(new_seq)
    # print("---")
    new_seqs.append(new_seq)

cc_phosphosites_to_a["Mutant Sequence"] = new_seqs
cc_phosphosites_to_a["Protein Change"] = prot_changes
cc_phosphosites_to_a["Mutation"] = "PhosphoSite Plus Phosphorylation Site -> A"
cc_phosphosites_to_a
```

```python
# Switching all phosphosites to Ds
cc_phosphosites_to_d = cc_phosphosites.copy(deep = True)
new_seqs = []
prot_changes2 = []

for i in cc_phosphosites_to_d.index:
    row = cc_phosphosites_to_d.loc[i]
    site_AD_pos = row.pos - row.Start
    orig_seq = row.ProteinRegionSeq
    new_seq = orig_seq[:site_AD_pos] + "d" + orig_seq[site_AD_pos + 1:]

    aa_dict = {"S" : "Ser",
               "T" : "Thr",
               "Y": "Tyr"}
    prot_changes2.append("p." + aa_dict[orig_seq[site_AD_pos]] + str(row.pos) + "Asp")

    print(row.MOD_RSD)
    print(orig_seq[site_AD_pos])
    print(orig_seq)
    print(new_seq)
    print("---")
    new_seqs.append(new_seq)

cc_phosphosites_to_d["Mutant Sequence"] = new_seqs
cc_phosphosites_to_d["Protein Change"] = prot_changes2
cc_phosphosites_to_d["Mutation"] = "PhosphoSite Plus Phosphorylation Site -> D"
cc_phosphosites_to_d
```

```python
cc_phosphosite_mutations = pd.concat([cc_phosphosites_to_a,cc_phosphosites_to_d])
cc_phosphosite_mutations
```

```python
cc_phosphosite_mutations.columns
```

```python
cc_phosphosite_mutations_to_save = cc_phosphosite_mutations[["Gene Name", "Mutant Sequence", "Protein Change", "Mutation"]].reset_index(drop = True)
cc_phosphosite_mutations_to_save
```

```python
cc_phosphosite_mutations_to_save.to_csv("../output/caitlin_experiment/phosphosite_plus.csv")
```
