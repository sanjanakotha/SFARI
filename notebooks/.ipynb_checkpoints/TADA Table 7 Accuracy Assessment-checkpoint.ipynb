### 1a. Assume GSL + Soto + ~~<mark>Stanford</mark>~~  represent real truth
"First, we assumed that GSL or Soto lists represented the real truth (which they obviously do not). Based on this assumption the TP = FP = etc, indicative of very poor performance"

```python
def return_tp_fp(preds_input, knownADs):
    preds = preds_input.copy()
    AD_comparison_tools.add_col_contains_prediction(preds, knownADs, result_col_name = "KnownAD")
    TP_count = sum(preds["KnownAD"])
    FP_count = sum(preds["KnownAD"] == False)
    print("True positives: " + str(TP_count) + "(" + str(np.round((100 * TP_count/ len(preds)), 1)) + "%)")
    print("False positives: " + str(FP_count)+ "(" + str(np.round((100 * FP_count/ len(preds)), 1)) + "%)")
    return TP_count, FP_count
```

```python
improved_unaggreg
```

```python
improved_unaggreg["pred"] = 1
improved_unaggreg
```

```python
tiled_lambert = pd.read_csv("../data/TilingDFs/LambertTFs_size_39_space_1_AAs_W,F,Y,L.csv", index_col = 0)
not_pred_tiles = tiled_lambert[~tiled_lambert.index.isin(improved_unaggreg.index)]
not_pred_tiles["pred"] = 0
not_pred_tiles
```

```python
all_tiles_pred_status = pd.concat([improved_unaggreg, not_pred_tiles])
all_tiles_pred_status
```

```python
all_tiles_pred_status["uniprotID"] = all_tiles_pred_status["GeneName"].str.split("|").str[1]
all_tiles_pred_status = all_tiles_pred_status.rename(columns = {"StartPosition" : "Start", "EndPosition" : "End"})
all_tiles_pred_status
```

```python
y_true = all_tiles_pred_status["known_AD"]
y_pred = all_tiles_pred_status["pred"]

f1_score(y_true, y_pred)
```
