```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import Levenshtein as lev
import itertools
```

```python
cc_names = pd.read_excel("../data/gene_names_with_location_and_pmvs.xlsx")
cc_names = cc_names.rename(columns = {"gene" : "Gene", "start" : "Start", "end" : "End"})
known_ADs = pd.read_csv("../output/known_ADs_considering_isoforms_and_canonical.csv")
cc_names = pd.merge(known_ADs[["uniprotID", "Gene", "Start", "End"]], cc_names)
ensts = pd.read_csv("../data/SFARI_TFs_with_ENST_corrected.csv", index_col = 0)
cc_names = pd.merge(cc_names, ensts)
cc_names["ENST"] = cc_names["ENST"].str.split(".").str[0]
cc_names["len"] = cc_names["End"] - cc_names["Start"] + 1
cc_names = cc_names.rename(columns = {"Start" : "start", "End" : "end"})
cc_names
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
import numpy as np

# Set the context for Seaborn
sns.set_context('paper')

# Open a PdfPages object to save the plots
with PdfPages('AD_tracks_all_subplots.pdf') as pdf:
    
    def plot_all_AD_tracks():
        num_plots = 30 + 14  # 2 * 15 + (15 - 1) Number of genes to plot (adjust as necessary)
        
        # Create a figure with 15 rows and 1 column
        fig, axes = plt.subplots(nrows=num_plots, ncols=1, figsize=(8.5, 11))

        # Flatten axes array for easier indexing
        axes = axes.flatten()

        # Loop over each gene and plot the data
        for idx, AD in enumerate(cc_names["Gene Name"]):  # Take first 15 genes
            #print(idx)

            
            AD_pt_vars = pt_vars[pt_vars["Gene Name"] == AD]
            AD_gnomad = gnomad[gnomad["Gene Name"] == AD]

            # Calculate the min and max position for x-axis
            min_pos = min(AD_pt_vars["pos"])
            max_pos = max(AD_pt_vars["pos"])

            # Get the current axis (ax1 is the top plot, ax2 is the bottom plot)
            ax1 = axes[3 * idx]  # Top plot for the AD data
            ax2 = axes[3 * idx + 1]  # Bottom plot for the gnomAD data

            # Plot AD data on the top plot (ax1)
            sns.histplot(data=AD_pt_vars, stat="count", multiple="stack",
                         x="pos", kde=False, hue="list", element="bars", legend=False,
                         bins=np.arange(min_pos, max_pos + 1, 1), ax=ax1)
            
            # Plot gnomAD data on the bottom plot (ax2)
            sns.histplot(data=AD_gnomad, stat="count", multiple="stack",
                         x="pos", kde=False, color="gray", element="bars", legend=True,
                         bins=np.arange(min_pos, max_pos + 1, 1), ax=ax2)

            # Adjust the y-axis limits for both subplots
            ax1_ymin, ax_1ymax = ax1.get_ylim()
            ax2_ymin, ax_2ymax = ax2.get_ylim()
            ymin, ymax = min(ax1_ymin, ax2_ymin), max(ax_1ymax, ax_2ymax)

            ax1.set_ylim(ymin, ymax)
            ax2.set_ylim(ymin, ymax)

            # Invert the y-axis for the bottom plot (ax2)
            ax2.invert_yaxis()
            ax2.set_ylabel("        Count")
            ax2.set_xticks(np.arange(min_pos, max_pos + 1, 20))
            ax2.set_xlim(min_pos, max_pos)
            ax2.set_xlabel("")

            # Set title for both subplots
            ax1.set_title(f"{AD} - Original Data")
            ax2.set_title(f"{AD} - gnomAD Data")

            # Remove spines
            sns.despine(ax=ax1)
            sns.despine(ax=ax2)

        # Adjust layout to prevent overlap between subplots
        plt.tight_layout()
        #plt.subplots_adjust(hspace=0)
        plt.subplots_adjust(hspace=0)
       
        # Save the current page (all plots) to the PDF
        pdf.savefig(fig)
        plt.close(fig)

    # Sort cc_names by the length of the gene names (optional)
    cc_names = cc_names.sort_values(by="len", ascending=True)

    # Plot and save all the tracks as subplots on a single page of the PDF
    plot_all_AD_tracks()

print("All plots have been saved to AD_tracks_all_subplots.pdf")
```
