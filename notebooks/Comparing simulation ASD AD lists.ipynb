```python
import pandas as pd
```

```python
new_known_ADs = pd.read_csv("../output/redone_ASD_assoc_known_ADs_on_TFs.csv")
new_known_ADs["Length"] = new_known_ADs["End"] - new_known_ADs["Start"] + 1
new_known_ADs
```

```python
old_SFARI_ADs = pd.read_csv("../output/SFARI_TF_known_ADs.csv", index_col = 0)
old_SFARI_ADs
```

```python
pd.merge(left = new_known_ADs, right = old_SFARI_ADs, on = ["Start", "End", "uniprotID"])
```

```python
new_known_ADs_filtered = new_known_ADs[["uniprotID", "Start", "End"]]
# new_known_ADs_filtered["list"] = "new"
old_SFARI_ADs_filtered = old_SFARI_ADs[["uniprotID", "Start", "End"]]
# old_SFARI_ADs_filtered["list"] = "old"
```

```python
pd.concat([new_known_ADs_filtered,old_SFARI_ADs_filtered]).drop_duplicates(keep=False)
```

```python
new_known_ADs["id"] = new_known_ADs["uniprotID"] + "|" + new_known_ADs["Start"].astype(str) + "|" + new_known_ADs["End"].astype(str)
new_known_ADs
```

```python
old_SFARI_ADs["id"] = old_SFARI_ADs["uniprotID"] + "|" + old_SFARI_ADs["Start"].astype(str) + "|" + old_SFARI_ADs["End"].astype(str)
old_SFARI_ADs
```

```python
new_ids = set(new_known_ADs["id"]) - set(old_SFARI_ADs["id"])
new_ids
```

```python
# New things to simulate
changed_simulation_ADs = new_known_ADs[new_known_ADs["id"].isin(new_ids)]
changed_simulation_ADs
```
