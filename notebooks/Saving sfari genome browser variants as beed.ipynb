```python
import pandas as pd
import glob
```

```python
pd.read_csv("../soto_analysis/raw_files/iWES_v2_variants.bed", nrows = 3, sep = "\t")
```

```python
sfari_genome_vars = []
for path in glob.glob("../output/sfari_json/*"):
    sfari_genome_vars.append(pd.read_csv(path))
sfari_genome_vars = pd.concat(sfari_genome_vars)
sfari_genome_vars
```

```python
sfari_genome_vars[0] = sfari_genome_vars["variant_id"].str.split("-").str[0]
sfari_genome_vars[2] = sfari_genome_vars["variant_id"].str.split("-").str[1]
sfari_genome_vars[1] = sfari_genome_vars[2].astype(int) - 1

sfari_genome_vars[3] = sfari_genome_vars["variant_id"].str.split("-").str[2]
sfari_genome_vars[4] = sfari_genome_vars["variant_id"].str.split("-").str[3]
sfari_genome_vars[5] = 0.5
sfari_genome_vars = sfari_genome_vars[[0, 1, 2, 3, 4, 5]]
sfari_genome_vars
```

```python
sfari_genome_vars.to_csv("../soto_analysis/raw_files/sfari_genome_vars.bed", sep = "\t", header = None, index = None)
```
