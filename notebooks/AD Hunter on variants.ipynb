```python
input_dict = protfasta.read_fasta("../../SFARI/soto_analysis/data/long_variant_ADs.fasta")
input_df = pd.DataFrame({"id" : input_dict.keys(), "sequence" : input_dict.values()})
input_df
```

```python
adhunter_preds = pd.read_csv("../data/adhunter_preds.csv", index_col = 0, converters={arr:pd.eval for arr in ['adhunter_centers', 'adhunter_preds']})
adhunter_preds
```

```python
adhunter_preds_with_id = pd.merge(adhunter_preds, input_df, how = "left", on = "sequence")
adhunter_preds_with_id
```

```python
def ranges_above_threshold(arr, threshold, min_length):
    ranges = []
    start = None
    for i, value in enumerate(arr):
        if value > threshold:
            if start is None:
                start = i
        elif start is not None:
            if i - start >= min_length:
                ranges.append((start, i - 1))
            start = None
    if start is not None and len(arr) - start >= min_length:
        ranges.append((start, len(arr) - 1))
    return ranges
```

```python
coords = []
for i in adhunter_preds_with_id.index:
    coords.append(ranges_above_threshold(adhunter_preds_with_id["adhunter_preds"].loc[i], 100000, 2))
adhunter_preds_with_id["AD_coords"] = coords
adhunter_preds_with_id
```

```python
expanded_active_adhunter_preds = adhunter_preds_with_id.explode("AD_coords").dropna()
expanded_active_adhunter_preds
```

```python
expanded_active_adhunter_preds["AD"] = expanded_active_adhunter_preds["id"].str.split("|").str[0]
expanded_active_adhunter_preds
```

```python
def active(arr):
    return max(arr) >= 100000
def add_active_col(df):
    activities = []
    for i in df.index:
        activities.append(active(df["adhunter_preds"].loc[i]))
    df["active"] = activities
```

```python
add_active_col(adhunter_preds_with_id)
adhunter_preds_with_id["AD"] = adhunter_preds_with_id["id"].str.split("|").str[0]
adhunter_preds_with_id
```

```python
add_active_col(wt_adhunter_preds_with_id)
wt_adhunter_preds_with_id
```

```python
active_wt_adhunter_preds_with_id = wt_adhunter_preds_with_id[wt_adhunter_preds_with_id["active"]]
active_wt_adhunter_preds_with_id
```

```python
active_wt_var_preds = adhunter_preds_with_id[adhunter_preds_with_id["AD"].isin(active_wt_adhunter_preds_with_id["id"])]
active_wt_var_preds
```

```python
preds_per_var = pd.DataFrame(active_wt_var_preds["AD"].value_counts()).reset_index()
preds_per_var
```

```python
input_df["AD"] = input_df["id"].str.split("|").str[0]
input_df
```

```python
number_vars = pd.DataFrame(input_df[input_df["AD"].isin(active_wt_adhunter_preds_with_id["id"])]["AD"].value_counts()).reset_index()
number_vars
```

```python
merged_df = pd.merge(number_vars, preds_per_var, on = "index")
merged_df = merged_df.rename(columns = {"AD_x" : "Number of variants",
                                       "AD_y" : "Number of predictions in variants"})
```
