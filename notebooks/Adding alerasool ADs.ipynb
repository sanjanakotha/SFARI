```python
#https://stackoverflow.com/questions/49071081/merging-overlapping-intervals-in-python
def return_merged_intervals(intervals):
    sorted_on_start = sorted(intervals)
    return recursive_merge(sorted_on_start.copy())

def recursive_merge(inter, start_index = 0):
    for i in range(start_index, len(inter) - 1):
        if inter[i][1] > inter[i+1][0]:
            new_start = inter[i][0]
            new_end = inter[i+1][1]
            inter[i] = [new_start, new_end]
            del inter[i+1]
            return recursive_merge(inter.copy(), start_index=i)
    return inter    

```

```python
merged_df = pd.DataFrame(columns = ["Gene", "uniprotID", "ProtSeq", "Merged Intervals"])

for gene in set(alerasool_AD_hits["Gene"]):
    gene_df = alerasool_AD_hits[alerasool_AD_hits["Gene"] == gene]
    #display(gene_df)

    intervals = [(s, e) for s, e in zip(gene_df["start"], gene_df["end"])]
    merged = return_merged_intervals(intervals)
    # print(intervals)
    # print(merged)
    # print('----\n')

    row = gene_df.iloc[0]
    merged_df.loc[len(merged_df)] = [row["Gene"], row["uniprotID"], row["ProtSeq"], str(merged)]

merged_df
```

```python
merged_df_exploded = merged_df.explode("Merged Intervals")
merged_df_exploded
```

```python
merged_df_exploded["Start"] = merged_df_exploded["Merged Intervals"].str.extract(r'(\d+),').astype(int)
merged_df_exploded["End"] = merged_df_exploded["Merged Intervals"].str.extract(r', (\d+)').astype(int)
merged_df_exploded
```

```python
merged_df_exploded["Sequence"] = [prot_seq[s-1:e] for s, e, prot_seq in zip(merged_df_exploded["Start"], merged_df_exploded["End"], merged_df_exploded["ProtSeq"])]
merged_df_exploded["len"] = merged_df_exploded["Sequence"].str.len()
merged_df_exploded
```

```python
alerasool_ADs_final = merged_df_exploded[["Gene", "uniprotID", "Start", "End", "Sequence"]]
alerasool_ADs_final["Reference"] = "Alerasool et al"
alerasool_ADs_final
```

```python
new_AD_count = 0
new_AD = []

for i in range(len(alerasool_ADs_final)):   
    row = alerasool_ADs_final.iloc[i]
    gene_curr_known_ADs = curr_known_ADs[curr_known_ADs["uniprotID"] == row["uniprotID"]]
    
    if len(gene_curr_known_ADs) == 0:
        new_AD_count += 1
        new_AD.append(True)
        display(alerasool_ADs_final.iloc[[i]])
    else:
        alerasool_start = row["Start"]
        alerasool_end = row["End"]

        found = False
        
        for j in range(len(gene_curr_known_ADs)):
            AD_start = gene_curr_known_ADs.iloc[j]["Start"]
            AD_end = gene_curr_known_ADs.iloc[j]["End"]

        if alerasool_start <= AD_end and AD_start <= alerasool_end:
            found = True

        if not found:
            new_AD_count += 1
            new_AD.append(True)
            display(alerasool_ADs_final.iloc[[i]])
        else:
            new_AD.append(False)

    #print()
        
alerasool_ADs_final["new_AD"] = new_AD
alerasool_ADs_final
```
