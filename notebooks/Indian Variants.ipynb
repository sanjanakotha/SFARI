```python
import pandas as pd
```

```python
south_asian_vars = pd.read_excel("../data/kerdoncuff_2024/Extended Data Table S1 Summary of loss of function (LOF) and missense variants in LASI-DAD.xlsx")
south_asian_vars
```

```python
south_asian_vars["chr"] = south_asian_vars["#ID"].str.split(":").str[0].astype(int)
south_asian_vars["end"] = south_asian_vars["#ID"].str.split(":").str[1].astype(int)
south_asian_vars["start"] = south_asian_vars["end"] - 1
south_asian_vars
```

```python
missense_south_asian_vars = south_asian_vars[south_asian_vars["Function"] == "missense"]
missense_south_asian_vars
```

```python
missense_south_asian_vars["ref_len"] = missense_south_asian_vars["REF"].str.len()
missense_south_asian_vars["alt_len"] = missense_south_asian_vars["REF"].str.len()
missense_south_asian_vars_snp = missense_south_asian_vars[(missense_south_asian_vars["ref_len"] == 1) & (missense_south_asian_vars["alt_len"] == 1)]
missense_south_asian_vars_snp
```

```python
missense_south_asian_vars_snp[["chr", "start", "end", "REF", "ALT", "Allele Frequency (AF)_LASI_DAD"]]
```

```python
missense_south_asian_vars_snp[["chr", "start", "end", "REF", "ALT", "Allele Frequency (AF)_LASI_DAD"]].to_csv("../soto_analysis/raw_files/south_asian.bed", index=False, sep='\t', header=None)
```
