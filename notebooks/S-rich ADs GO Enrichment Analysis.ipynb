```python
import pandas as pd
```

```python
known_ADs = pd.read_csv("../output/known_ADs_considering_isoforms.csv")
known_ADs
```

```python
known_ADs["S"] = known_ADs["ProteinRegionSeq"].str.count("S")
known_ADs["len"] = known_ADs["ProteinRegionSeq"].str.len()

known_ADs["S_prop"] = known_ADs["S"] / known_ADs["len"]
known_ADs
```

```python
S_rich_ADs = known_ADs[known_ADs["S_prop"] > 0.15]
S_rich_ADs
```

```python
S_rich_AD_uniprotIDs = S_rich_ADs[["uniprotID"]]
S_rich_AD_uniprotIDs = S_rich_AD_uniprotIDs.drop_duplicates()
S_rich_AD_uniprotIDs.to_csv("../data/S_rich_AD_uniprotIDs.csv", index = False, header = None)
S_rich_AD_uniprotIDs
```
