```python
break_indices = [-1] + list(maf[maf[0] == "a"].index)
break_indices
```

```python
merged_df = maf[maf[0] == "s"][[0, 1]].drop_duplicates()

for i in range(len(break_indices) - 1):
    start = break_indices[i]
    end = break_indices[i+1]
    df = maf.loc[start + 1:end - 1]
    df = df.rename(columns = {6 : "seq_" + str(i)})
    #display(merged_df)
    merged_df = pd.merge(merged_df, df, on = [0, 1], how = "left")

merged_df
```
