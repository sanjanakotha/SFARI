```python
AD_1_variant_dict = protfasta.read_fasta("../outputs/AD_variant_fasta/VDR_P11473_AD_195-238")
AD_1 = pd.DataFrame({"id" : AD_1_variant_dict.keys(),
                    "seq" : AD_1_variant_dict.values()})
AD_1
```

```python
AD_1["full_prot_pos"] = AD_1["id"].str.extract('.*(\d\d\d).*')
AD_1
```

```python
AD_1["domain_pos"] = AD_1["full_prot_pos"].astype(int) - 195 + 1
AD_1
```

```python
AD_1["orig"] = AD_1["id"].str.extract('.*(...)\d\d\d.*')
AD_1["orig"] =AD_1["orig"].str.upper()
AD_1["var"] = AD_1["id"].str.extract('.*\d\d\d(...).*')
AD_1["var"] =AD_1["var"].str.upper()
AD_1
```

```python
d = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
     'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
     'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
     'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M'}
```

```python
AD_1["var"] = AD_1["var"].replace(d)
AD_1["orig"] = AD_1["orig"].replace(d)
```

```python
AD_1
```

```python
# The sequence in the plasmid actually does not have the first 3 positions, so adjusting
AD_1["seq_plasmid"] = AD_1["seq"].str[3:]
AD_1["domain_pos_plasmid"] = AD_1["domain_pos"] - 3
AD_1
```

```python
AD_1_plasmid_df = AD_1[AD_1["domain_pos_plasmid"] > 0]
AD_1_plasmid_df
```

```python
AD_1_plasmid_df["mutation"] = AD_1_plasmid_df["orig"].astype(str) + AD_1_plasmid_df["domain_pos_plasmid"].astype(str) + AD_1_plasmid_df["var"].astype(str)
AD_1_plasmid_df
```

```python
" ".join(AD_1_plasmid_df["mutation"])
```
