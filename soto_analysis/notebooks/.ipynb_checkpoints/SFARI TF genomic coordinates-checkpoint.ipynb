```python
genomic_coords = pd.read_csv("../outputs/mart_export_tf_genomic_coords.txt", sep = "\t")
genomic_coords
```

```python
ENST_codes = set(genomic_coords["Transcript stable ID"])
ENST_codes
```

```python
len(ENST_codes)
```

```python
for ENST_code in list(ENST_codes):
    print(ENST_code)
    df = genomic_coords[genomic_coords["Transcript stable ID"] == ENST_code].dropna()
    df = df[["Chromosome/scaffold name", "Genomic coding start",
            "Genomic coding end", "Transcript stable ID", "Strand"]]
    df["Genomic coding start"] = df["Genomic coding start"] - 1
    df["Genomic coding start"]= df["Genomic coding start"].astype(int)
    df["Genomic coding end"]= df["Genomic coding end"].astype(int)
    df.to_csv("../outputs/mutations/cds_bed_format/"+ENST_code+".bed", index=False, sep='\t', header=None)
```
