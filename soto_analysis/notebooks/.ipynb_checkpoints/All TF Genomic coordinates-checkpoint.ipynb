```python
genomic_coords = pd.read_csv("../../data/mart_export_all_tf_genomic_coords.txt", sep = "\t")
genomic_coords
```

```python
genomic_coords["Chromosome/scaffold name"].value_counts()
```

```python
accep_chrs = list(genomic_coords["Chromosome/scaffold name"].value_counts().index[:23]) + ["Y"]
accep_chrs
```

```python
ENST_codes = set(genomic_coords["Transcript stable ID"])
len(ENST_codes)
```

```python
issues = []
for ENST_code in list(ENST_codes):

#for ENST_code in ["ENST00000056233"]:
    #print(ENST_code)
    df = genomic_coords[genomic_coords["Transcript stable ID"] == ENST_code].dropna()
    df = df[["Chromosome/scaffold name", "Genomic coding start",
            "Genomic coding end", "Transcript stable ID", "Strand"]]

    # Drop any rows with Human genome issues
    keep = df[df["Chromosome/scaffold name"].isin(accep_chrs)]

    if len(keep) > 0:
        df["Genomic coding start"] = df["Genomic coding start"]
        df["Genomic coding start"]= df["Genomic coding start"].astype(int) - 1
        df["Genomic coding end"]= df["Genomic coding end"].astype(int)
        #display(df)
        df.to_csv("../outputs/mutations/cds_bed_format/"+ENST_code, index=False, sep='\t', header=None)
    else:
        issues.append(ENST_code)
        print(ENST_code)
```
