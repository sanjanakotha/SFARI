```python
import pandas as pd
import protfasta
import re
```

```python
from os import listdir
from os.path import isfile, join
#https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
ADs = [f for f in listdir("../outputs/AD_variant_fasta/") if isfile(join("../outputs/AD_variant_fasta/", f))]
```

```python
def return_AD_df(AD):
    if "KMT2A-KMT2A-ALL1-HRX-KMT2A-MLL1-CXXC7" not in AD: #There is an issue with the coordinates for this, so skipping for now
        AD_dict = protfasta.read_fasta("../outputs/AD_variant_fasta/" + AD)
        df = pd.DataFrame({"name" : AD_dict.keys(), "seq" : AD_dict.values()})
        df["name"] = AD + "|" + df["name"]
        # print(AD)
        # display(df)
        return df
```

```python
dfs_to_concat = []
for AD in ADs:
    dfs_to_concat.append(return_AD_df(AD))
```

```python
all_variant_ADs = pd.concat(dfs_to_concat)
all_variant_ADs
```

```python
all_variant_ADs["len"] = all_variant_ADs["seq"].str.len()
all_variant_ADs
```

```python
long_all_variant_ADs = all_variant_ADs[all_variant_ADs["len"] >= 53]
long_all_variant_ADs
```

```python
# Also adding the wt sequences for each
```

```python
known_ADs = pd.read_csv("../../output/known_ADs_considering_isoforms_and_canonical.csv")
known_ADs
```

```python
long_all_variant_ADs["uniprotID"] = long_all_variant_ADs["name"].str.split("_").str[1]
long_all_variant_ADs["start_to_end"] = long_all_variant_ADs["name"].str.split("_").str[-1]
long_all_variant_ADs["start_to_end"] = long_all_variant_ADs["start_to_end"].str.split("|").str[0]
long_all_variant_ADs["Start"] = long_all_variant_ADs["start_to_end"].str.split("-").str[0].astype(int)
long_all_variant_ADs["End"] = long_all_variant_ADs["start_to_end"].str.split("-").str[1].astype(int)
long_all_variant_ADs
```

```python
wt_long_variant_AD_coords = long_all_variant_ADs[["uniprotID", "Start", "End"]].drop_duplicates().reset_index(drop = True)
wt_long_variant_AD_coords
```

```python
wt_long_variant_ADs = pd.merge(wt_long_variant_AD_coords, known_ADs, how = "left", on = ["uniprotID", "Start", "End"])
wt_long_variant_ADs
```

```python
wt_long_variant_ADs["name"] = wt_long_variant_ADs["Gene"] + "_" + wt_long_variant_ADs["uniprotID"] + "_" + "AD" + "_" + wt_long_variant_ADs["Start"].astype(str) + "-" + wt_long_variant_ADs["End"].astype(str)
wt_long_variant_ADs["seq"] = wt_long_variant_ADs["ProteinRegionSeq"]
wt_long_variant_ADs
```

```python
long_all_variant_ADs_dict = dict(zip(long_all_variant_ADs["name"], long_all_variant_ADs["seq"]))
```

```python
protfasta.write_fasta(long_all_variant_ADs_dict, '../data/long_variant_ADs.fasta')
```

```python
wt_ADs_dict = dict(zip(wt_long_variant_ADs["name"], wt_long_variant_ADs["seq"]))
```

```python
protfasta.write_fasta(wt_ADs_dict, '../data/wt_ADs.fasta')
```
