```python
import pandas as pd
```

```python
all_TFs_table = pd.read_csv("../outputs/all_TFs_table_proteins.txt", sep = "\t", index_col = 0)
```

```python
def display_diff_summary(uniprotID):
    # Get enst used
    ENST = all_TFs_table[all_TFs_table["uniprotID"] == uniprotID]["ENST"].iloc[0]

    # 1. CDS coordinates comparison
    print("CDS COORDINATES\n")
    print("Old")
    display(pd.read_csv("../outputs/mutations_old2/cds_bed_format/" + uniprotID + ".bed", sep = "\t", header = None).sort_values(by = 1))
    
    print("New")
    display(pd.read_csv("../outputs/mutations/cds_bed_format/" + ENST, sep = "\t", header = None).sort_values(by = 1))

    print('─' * 100)    
    
    # 2. CDS variants 
    print("CDS VARIANTS\n")
    print("Old")
    display(pd.read_csv("../outputs/mutations_old2/cds_expanded_iWES_v2_variants/" + uniprotID + ".bed", sep = "\t", header = None))

    print("New")
    display(pd.read_csv("../outputs/mutations/cds_expanded_iWES_v2_variants/" + ENST + ".bed", sep = "\t", header = None))

    print('─' * 100)    
    
    # 3. CDS Variants classified

    print("CDS VARIANTS CLASSIFIED\n")
    print("Old")
    display(pd.read_csv("../outputs/mutations_old2/cds_expanded_iWES_v2_variants_snv_classified/" + uniprotID + ".bed", sep = "\t", header = None))

    print("New")
    display(pd.read_csv("../outputs/mutations/cds_expanded_iWES_v2_variants_snv_classified/" + ENST + ".bed", sep = "\t", header = None))
```

```python
display_diff_summary("Q92731")
```

```python
display_diff_summary("Q13422")
```

```python
display_diff_summary("Q9UGL1")
```
