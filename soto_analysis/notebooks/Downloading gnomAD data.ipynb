```python
import os

# Directory path
directory = '../raw_files/gnomAD/'

# Get all files in the directory
gnomAD_files = [file for file in os.listdir(directory) if os.path.isfile(os.path.join(directory, file))]

# Print the list of files
print("Files in the directory:")
for file in gnomAD_files:
    print(file)
```

```python
dfs = []

for file in gnomAD_files:
    if file[0] == "g":
        output_df = pd.read_csv("../raw_files/gnomAD/" + str(file), low_memory = False)
        output_df["start"] = output_df["Position"] - 1
        output_df = output_df[["Chromosome", "start", "Position", "Reference", "Alternate", "Allele Frequency"]]
        dfs.append(output_df)
```

```python
all_gnomad_variants = pd.concat(dfs)
all_gnomad_variants
```

```python
ref_len = all_gnomad_variants["Reference"].str.len() == 1
alt_len = all_gnomad_variants["Alternate"].str.len() == 1
```

```python
gnomad_snps = all_gnomad_variants[ref_len & alt_len]
gnomad_snps
```

```python
gnomad_snps.to_csv("../raw_files/gnomad.bed", index=False, sep='\t', header=None)
```
