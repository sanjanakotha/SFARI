```python
import pandas as pd
```

```python
all_AD_vs_TF_hits = set(["MEIS2", "RORA", "CASZ1", "YY1", "NKX2-2", 
                       "TCF7L2", "NR4A2", "NCOA1"])
rare_AD_vs_TF_hits = set(["MEIS2", "RORA", "TCF7L2", "YY1", "NKX2-2",
                          "NCOA1", "NR4A2", "IKZF1", "CASZ1"])
```

```python
all_AD_vs_DBD_hits = set(["MEIS2", "NKX2-2", "ERG", "PAX6", "CAMTA2",
                            "PAX5", "PITX1", "ARX", "OTX1", "NR1D1"])
rare_AD_vs_DBD_hits = set(["ARX", "CAMTA2", "ERG", "MEIS2", "NKX2-2",
                          "OTX1", "PAX5", "PAX6", "PITX1"])
```

```python
all_AD_vs_TF_hits.intersection(rare_AD_vs_TF_hits, all_AD_vs_DBD_hits, rare_AD_vs_DBD_hits)
```

```python
all_hits = all_AD_vs_TF_hits.union(rare_AD_vs_TF_hits).union(all_AD_vs_DBD_hits).union(rare_AD_vs_DBD_hits)
all_hits
```

```python
all_hits_df = pd.DataFrame({"GeneName" : list(all_hits)})
all_hits_df
```

```python
all_hits_df["all_AD_vs_TF"] = all_hits_df["GeneName"].isin(all_AD_vs_TF_hits).astype(int)
all_hits_df["rare_AD_vs_TF"] = all_hits_df["GeneName"].isin(rare_AD_vs_TF_hits).astype(int)
all_hits_df["all_AD_vs_DBD"] = all_hits_df["GeneName"].isin(all_AD_vs_DBD_hits).astype(int)
all_hits_df["rare_AD_vs_DBD"] = all_hits_df["GeneName"].isin(rare_AD_vs_DBD_hits).astype(int)
```

```python
all_hits_df
```

```python
all_hits_df["hit_count"] = all_hits_df["all_AD_vs_TF"] \
        + all_hits_df["rare_AD_vs_TF"] \
        + all_hits_df["all_AD_vs_DBD"] \
        + all_hits_df["rare_AD_vs_DBD"]
all_hits_df
```

```python
all_hits_df.sort_values(by = "hit_count", ascending = False)
```

```python
all_hits_df.sort_values(by = "hit_count", ascending = False).to_csv("../outputs/hit_summary.csv")
```
