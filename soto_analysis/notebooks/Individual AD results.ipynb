```python
%load_ext autoreload

%autoreload 2
import pandas as pd
import numpy as np
%aimport Variant_analysis_helper_functions


import protfasta
import os
```

```python
filepath = "../outputs/AD_variant_fasta/"
ADs = [f for f in os.listdir(filepath) if os.path.isfile(os.path.join(filepath, f))]
ADs
```

```python
num_vars = []

for AD in ADs:
    print(AD)
    AD_fasta_dict = protfasta.read_fasta("../outputs/AD_variant_fasta/" + AD)
    num_vars.append((len(AD_fasta_dict)))
```

```python
AD_df = pd.DataFrame({"AD" : ADs, "AD_missense" : num_vars})
AD_df
```

```python
AD_df["GeneName"] = AD_df["AD"].str.split("_").str[0]
AD_df["uniprotID"] = AD_df["AD"].str.split("_").str[1]
AD_df["location"] = AD_df["AD"].str.split("_").str[-1]
AD_df["start"] = AD_df["location"].str.split("-").str[0].astype(int)
AD_df["end"] = AD_df["location"].str.split("-").str[1].astype(int)
AD_df["AD_cds_length"] = 3 * (AD_df["end"] - AD_df["start"] + 1)
AD_df["AD_missense_prop"] = AD_df["AD_missense"] / AD_df["AD_cds_length"]
AD_df
```

```python
DBD_results_tbl = pd.read_csv("../outputs/DBD_results_tbl.csv", index_col = 0)
DBD_results_tbl = DBD_results_tbl[["gene-symbol", "uniprotID", 
                                   "DBD_cds_length", "DBD_missense", "DBD_missense_prop",
                                  "TF_cds_length", "TF_missense", "TF_missense_prop"]]
DBD_results_tbl
```

```python
AD_and_DBDs = pd.merge(DBD_results_tbl, AD_df, how = "left", on = "uniprotID")
AD_and_DBDs
```

```python
Variant_analysis_helper_functions.add_fisher_p_vals_vs_control(AD_and_DBDs, "AD", "DBD")
```
